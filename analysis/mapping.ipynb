# Mapping Feature
- API using nominatim
- Folium map with OSM

```python
import pandas as pd
import geopandas as gpd
from geopy.geocoders import Nominatim
import leafmap.foliumap as leafmap
```

```python
# testing geolocator library
geolocator = Nominatim(user_agent="u2004763@siswa.um.edu.my")
location = geolocator.geocode("Johor Bahru") # testing

print(location.point)
print(location.address)

# print out longitude and latitude
point = location.point
print("Latitude:", point.latitude)
print("Longitude:", point.longitude)
```

```python
# make the output as a dataframe
df = pd.DataFrame({'City': ['Johor Bahru'], 'Latitude': [point.latitude], 'Longitude': [point.longitude]})
df
```

# Mapping

```python
m = leafmap.Map(center=[df['Latitude'][0], df['Longitude'][0]], zoom=15, google_map="HYBRID")
m.add_points_from_xy(
    df, 
    x="Longitude", 
    y="Latitude",
    icon_names=['gear', 'map', 'leaf', 'globe'],
)
m
```
